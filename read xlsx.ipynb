```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data_PU = pd.read_excel('Dataset PU.xlsx')
data_PU.sort_values(["tanggal"], axis = 0, ascending=[True], inplace=True)
print(data_PU)

# Scatter Plot
#plt.figure(figsize=(10, 6))
plt.scatter(data_PU['tanggal'], data_PU['pm10'])
plt.title('Data Jenis Pollutant pm10')
plt.xlabel('tanggal')
plt.ylabel('jumlah pollutant')
plt.show()
```

```python
data_PU.sort_values(["categori"], axis = 0, ascending=[True], inplace=True)
print(data_PU)
plt.bar(data_PU['location'], data_PU['categori'])
plt.title('tanggal vs pm10')
plt.xlabel('tanggal')
plt.ylabel('jumlah pollutant')
plt.show()
```

```python
plt.plot(data_PU['tanggal'], data_PU['so2'])
plt.title('Data Jenis Pollutant so2')
plt.xlabel('tanggal')
plt.ylabel('jumlah pollutant')
plt.show()
```
