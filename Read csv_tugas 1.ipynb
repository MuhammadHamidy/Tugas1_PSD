```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data_PU = pd.read_csv('DatasetPU.csv')

# List of pollutants
pollutants = ['pm10', 'so2', 'co', 'o3', 'no2', 'max']

# Calculate mean, median, and mode for each pollutant at each location
stats = data_PU.groupby('location')[pollutants].agg(['mean', 'median', lambda x: x.mode().iloc[0] if not x.mode().empty else None])
stats.columns = [f'{pollutant}_{stat}' for pollutant in pollutants for stat in ['mean', 'median', 'mode']]

# Create subplots for each pollutant
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
fig.suptitle('Mean, Median, and Mode of Pollutants by Location', fontsize=16)

for i, pollutant in enumerate(pollutants):
    ax = axes[i // 3, i % 3]
    
    mean = stats[f'{pollutant}_mean']
    median = stats[f'{pollutant}_median']
    mode = stats[f'{pollutant}_mode']
    
    x = range(len(stats))
    
    ax.plot(x, mean, 'ro-', label='Mean')
    ax.plot(x, median, 'go-', label='Median')
    ax.plot(x, mode, 'bo-', label='Mode')
    
    ax.set_title(f'{pollutant.upper()}')
    ax.set_xlabel('Location')
    ax.set_ylabel('Value')
    ax.set_xticks(x)
    ax.set_xticklabels(stats.index, rotation=45, ha='right')
    ax.legend()

plt.tight_layout()
plt.show()

# Print the statistics
print(stats)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Read the CSV file
data_PU = pd.read_csv('DatasetPU.csv')

# List of pollutants
pollutants = ['pm10', 'so2', 'co', 'o3', 'no2', 'max']

# Calculate mean, median, mode, kurtosis, and skewness for each pollutant at each location
stats_data = data_PU.groupby('location')[pollutants].agg(['mean', 'median', lambda x: x.mode().iloc[0] if not x.mode().empty else None, 'kurtosis', 'skew'])
stats_data.columns = [f'{pollutant}_{stat}' for pollutant in pollutants for stat in ['mean', 'median', 'mode', 'kurtosis', 'skew']]

# Create subplots for kurtosis and skewness
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
fig.suptitle('Kurtosis and Skewness of Pollutants by Location', fontsize=16)

for i, pollutant in enumerate(pollutants):
    ax = axes[i // 3, i % 3]
    
    kurtosis = stats_data[f'{pollutant}_kurtosis']
    skewness = stats_data[f'{pollutant}_skew']
    
    x = range(len(stats_data))
    
    ax.plot(x, kurtosis, 'ro-', label='Kurtosis')
    ax.plot(x, skewness, 'bo-', label='Skewness')
    
    ax.set_title(f'{pollutant.upper()}')
    ax.set_xlabel('Location')
    ax.set_ylabel('Value')
    ax.set_xticks(x)
    ax.set_xticklabels(stats_data.index, rotation=45, ha='right')
    ax.legend()

plt.tight_layout()
plt.show()

# Print the statistics
print(stats_data)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data_PU = pd.read_csv('DatasetPU.csv')

# List of pollutants
pollutants = ['pm10', 'so2', 'co', 'o3', 'no2', 'max']

# Function to calculate statistics for each pollutant
def calculate_stats(group):
    return pd.DataFrame({
        'mean': group.mean(),
        'median': group.median(),
        'mode': group.mode().iloc[0] if not group.mode().empty else np.nan,
        'kurtosis': group.kurtosis(),
        'skewness': group.skew()
    })

# Calculate statistics for each location and pollutant
stats_by_location = data_PU.groupby('location')[pollutants].apply(calculate_stats)

# Reshape the data for plotting
stats_melted = stats_by_location.reset_index()
stats_melted = pd.melt(stats_melted, id_vars=['location', 'level_1'], 
                       var_name='statistic', value_name='value')
stats_melted = stats_melted.rename(columns={'level_1': 'pollutant'})

# Set up the plot
plt.figure(figsize=(20, 15))
sns.set(style="whitegrid")

# Create subplots for each statistic
stats_to_plot = ['mean', 'median', 'mode', 'kurtosis', 'skewness']
for i, stat in enumerate(stats_to_plot, 1):
    plt.subplot(3, 2, i)
    sns.barplot(x='location', y='value', hue='pollutant', 
                data=stats_melted[stats_melted['statistic'] == stat])
    plt.title(f'{stat.capitalize()} by Location and Pollutant')
    plt.xticks(rotation=45)
    plt.legend(title='Pollutant', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('location_stats_analysis.png', dpi=300, bbox_inches='tight')
plt.close()

print("Analysis complete. The graph has been saved as 'location_stats_analysis.png'.")

# Print summary statistics
print("\nSummary Statistics:")
print(stats_by_location.to_string())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data_PU = pd.read_csv('DatasetPU.csv')

# List of pollutants
pollutants = ['pm10', 'so2', 'co', 'o3', 'no2', 'max']

# Function to calculate kurtosis and skewness for each pollutant
def calculate_kurt_skew(group):
    return pd.DataFrame({
        'kurtosis': group.kurtosis(),
        'skewness': group.skew()
    })

# Calculate kurtosis and skewness for each location and pollutant
kurt_skew_by_location = data_PU.groupby('location')[pollutants].apply(calculate_kurt_skew)

# Reshape the data for plotting
kurt_skew_melted = kurt_skew_by_location.reset_index()
kurt_skew_melted = pd.melt(kurt_skew_melted, id_vars=['location', 'level_1'], 
                           var_name='statistic', value_name='value')
kurt_skew_melted = kurt_skew_melted.rename(columns={'level_1': 'pollutant'})

# Set up the plot
plt.figure(figsize=(16, 10))
sns.set(style="whitegrid")

# Create subplots for kurtosis and skewness
plt.subplot(1, 2, 1)
sns.barplot(x='location', y='value', hue='pollutant', 
            data=kurt_skew_melted[kurt_skew_melted['statistic'] == 'kurtosis'])
plt.axhline(y=3, color='r', linestyle='--', label='Normal (Kurtosis=3)')
plt.title('Kurtosis by Location and Pollutant')
plt.xticks(rotation=45)
plt.legend(title='Pollutant', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.subplot(1, 2, 2)
sns.barplot(x='location', y='value', hue='pollutant', 
            data=kurt_skew_melted[kurt_skew_melted['statistic'] == 'skewness'])
plt.axhline(y=0, color='r', linestyle='--', label='Normal (Skewness=0)')
plt.title('Skewness by Location and Pollutant')
plt.xticks(rotation=45)
plt.legend(title='Pollutant', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('kurtosis_skewness_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("Kurtosis and Skewness graphs saved as 'kurtosis_skewness_analysis.png'.")
```
